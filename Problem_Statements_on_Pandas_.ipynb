```python
#Create a CSV file containing product names and prices. Read the data using
#Pandas. Use NumPy to calculate the average price. Write a function to identify
#products priced above the average and display them.

import pandas as pd
import numpy as np


d1 = pd.read_csv("/content/product1.csv")


d1["price"] = d1["price"].str.replace(",", "").astype(int)


avg_price = np.average(d1["price"].to_numpy())
print("Average Price:", avg_price)


def products_above_average(df, avg_price):
    return df[df["price"] > avg_price]


above_avg = products_above_average(d1, avg_price)
print("Products priced above average:\n", above_avg)
```

```python
#Read daily temperature data from a CSV file using Pandas. Use NumPy to
#calculate the highest, lowest, and average temperatures. Write functions to classify
#each day as:

d2 = pd.read_csv("/content/temperature.csv")


highest = np.max(d2["temperature"].to_numpy())
lowest = np.min(d2["temperature"].to_numpy())
average = np.mean(d2["temperature"].to_numpy())

print("Highest Temperature:", highest)
print("Lowest Temperature:", lowest)
print("Average Temperature:", average)


def classify(temp):
    if temp < 20:
        return "Cold"
    elif 20 <= temp <= 30:
        return "Moderate"
    else:
        return "Hot"


d2["Category"] = d2["temperature"].apply(classify)


print("\nDaily Temperature Classification:\n", d2)
```
